```python
import pandas as pd
import sys
from io import StringIO
from rdkit.Chem import AllChem, DataStructs, Descriptors, Descriptors3D
from rdkit import Chem
import numpy as np
```

```python
# read training data
data=pd.read_csv('Supp file S2 Result Table 20250904.txt',sep='\t', low_memory=False)
data.reset_index(drop=True, inplace=True)

columns_to_keep = ['ID', 'flat_smiles','final class']
data= data[columns_to_keep]
data=data.rename(columns={"ID": "SUBSTANCE"})
data=data.rename(columns={"flat_smiles": "SMILES_STD_FLAT"})
data=data.rename(columns={"final class": "class"})
print(data.shape[0])

data.head()
```

```python
def getRDKitDescriptors(df, mol_col='mol', smiles_col='SMILES'):
    if mol_col not in df.columns:
        df[mol_col] = df[smiles_col].apply(make_mol)
        #df = df[df[mol_col]!= 'remove']
    for descr in Descriptors._descList:
        df[descr[0]] = df[mol_col].apply(lambda x: x if x is None else descr[1](x))
    return df
```

```python
def make_mol(row, silent=True):
    # redirect stderr to silence rdkit
    if silent:
        io_orig = sys.stderr
        io = sys.stderr = StringIO()
    try:
        res = Chem.MolFromSmiles(row)
    except:
        res = None
    # restore stderr
    if silent:
        sys.stderr = io_orig
    return res
```

```python
data = getRDKitDescriptors(data, smiles_col='SMILES_STD_FLAT')
data = data.dropna(subset=['SUBSTANCE', 'SMILES_STD_FLAT'], axis=0)
print('shape of data: ', data.shape)
data = data.mask(data.eq('None')).dropna()
print('shape of data: ', data.shape)
data = data.drop_duplicates(subset=['SUBSTANCE', 'SMILES_STD_FLAT'])
print('shape of data: rdkit descriptors added, na SMILES, mol & IDs, duplicates removed', data.shape)
```

```python
def fp_as_array(mol, fingerprint_radius, fingerprint_bits):
    fp = AllChem.GetMorganFingerprintAsBitVect(mol, fingerprint_radius, nBits=fingerprint_bits)
    arr = np.zeros((1,), int)
    DataStructs.ConvertToNumpyArray(fp, arr)
    return arr
```

```python
def build_dataframe(df, smiles_col='SMILES_STD_FLAT'):
    df = df[[smiles_col]]
    df['mol'] = [Chem.MolFromSmiles(x) for x in df.SMILES_STD_FLAT]
    df['morgan'] = [fp_as_array(x, fingerprint_radius=3, fingerprint_bits=1024) for x in df.mol]
    return df
```

```python
# calculate Morgan Fingerprints
df_morgan = data.copy()
df_morgan = build_dataframe(df_morgan)
morgan_df = pd.DataFrame(df_morgan['morgan'].tolist(), index=df_morgan.index)
morgan_df.columns = [i for i in range(morgan_df.shape[1])]
df_expanded = pd.concat([df_morgan, morgan_df], axis=1)
df_expanded.drop(columns = ['morgan','mol'], inplace=True)
df = data.merge(df_expanded, on = 'SMILES_STD_FLAT')
print('shape of df after adding fp', df.shape)
```
